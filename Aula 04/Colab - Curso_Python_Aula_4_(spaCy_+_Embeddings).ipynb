```python
import spacy
from spacy import util as spc_util
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

palavras_similaridade = "madri espanha frança paris"

nlp = spacy.load("pt_core_news_lg")
doc = nlp(palavras_similaridade)
tokens = [token for token in doc]

pathw2v = '/content/drive/My Drive/arquivos_txt/vectors_spacy'
spc_util.load_model(pathw2v, vocab=nlp.vocab)

vetor_resultado = np.array(tokens[0].vector) - np.array(tokens[1].vector) + np.array(tokens[2].vector)

#É necessário remodelar o vetor retornado pelo spaCy, pois ele está em 1 dimensão e para o uso da similaridade do cosseno, é necessário um vetor de 2 dimensões!
vetor_resultado = vetor_resultado.reshape(1, -1)
vetor_paris = tokens[3].vector.reshape(1, -1)

similaridade = cosine_similarity(vetor_resultado, vetor_paris)
print(similaridade)
```
